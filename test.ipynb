```python
import numpy as np, matplotlib.pyplot as plt, paddle
from paddle.vision import transforms
from paddle.io import DataLoader
```

```python
ds = np.load('./shengxiao.npz')
images, labels = ds['images'], ds['labels']

mean, std = images.mean(axis=(0, 1, 2)), images.std(axis=(0, 1, 2))
transformer = transforms.Normalize(mean=mean, std=std, data_format='HWC')
rotator = transforms.RandomRotation(180)
```

```python
# 标准化并随机旋转图像
def transform_image_generator(images):
    for i in range(len(images)):
        normal_image = transformer(images[i])
        rotated_image = rotator(normal_image)
        yield [normal_image, rotated_image]

transform_images = []

for idx, img in enumerate(transform_image_generator(images)):
    transform_images.append(img[0])
    transform_images.append(img[1])

transform_images = np.array(transform_images)
# 扩展Labels
labels = np.array([val for val in labels for _ in range(2)])
```
